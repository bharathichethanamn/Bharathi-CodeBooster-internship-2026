```python
import pandas as pd
import sqlite3
df = pd.read_csv('/content/drive/MyDrive/student_performance.csv')
conn = sqlite3.connect("students.db")
df.to_sql('students', conn, if_exists='replace', index=False)

print("Database and table created successfully!")
```

```python
import pandas as pd

def run_query(query, title):
    print("\n" + "="*60)
    print(title)
    print("="*60)

    result = pd.read_sql_query(query, conn)

    print(result)

    return result
```

```python
query_avg_math_department = """
SELECT department,
       AVG(math_score) AS avg_math_score
FROM students
GROUP BY department;
"""

result_avg_math_department = run_query(
    query_avg_math_department,
    "Query: Average Math Score by Department"
)
```

```python

query_student_count_department = """
SELECT department,
       COUNT(*) AS student_count
FROM students
GROUP BY department;
"""

result_student_count_department = run_query(
    query_student_count_department,
    "Query: Student Count by Department"
)
```

```python

query_top_8_students = """
SELECT name,
       department,
       (math_score + programming_score + science_score) AS total_score
FROM students
ORDER BY total_score DESC
LIMIT 8;
"""

result_top_8_students = run_query(
    query_top_8_students,
    "Query: Top 8 Students by Total Score"
)
```

```python

query_gender_avg_attendance = """
SELECT gender,
       AVG(attendance_percentage) AS avg_attendance
FROM students
GROUP BY gender;
"""

result_gender_avg_attendance = run_query(
    query_gender_avg_attendance,
    "Query: Gender-wise Average Attendance"
)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

axes[0, 0].bar(
    result_avg_math_department['department'],
    result_avg_math_department['avg_math_score']
)

axes[0, 0].set_title("Average Math Score by Department")
axes[0, 0].set_xlabel("Department")
axes[0, 0].set_ylabel("Average Math Score")

axes[0, 1].pie(
    result_student_count_department['student_count'],
    labels=result_student_count_department['department'],
    autopct='%1.1f%%'
)

axes[0, 1].set_title("Student Count by Department")
axes[1, 0].barh(
    result_top_8_students['name'],
    result_top_8_students['total_score']
)

axes[1, 0].set_title("Top 8 Students by Total Score")
axes[1, 0].set_xlabel("Total Score")
axes[1, 0].set_ylabel("Student Name")
axes[1, 1].bar(
    result_gender_avg_attendance['gender'],
    result_gender_avg_attendance['avg_attendance']
)
axes[1, 1].set_title("Average Attendance by Gender")
axes[1, 1].set_xlabel("Gender")
axes[1, 1].set_ylabel("Average Attendance")
plt.tight_layout()
plt.show()
```

```python
query_avg_programming_female = """
SELECT AVG(programming_score) AS avg_programming_score
FROM students
WHERE gender = 'Female';
"""

result_avg_programming_female = run_query(
    query_avg_programming_female,
    "Query: Average Programming Score for Female Students"
)
```

```python
query_where_example = """
SELECT *
FROM students
WHERE attendance_percentage > 90;
"""

result_where_example = run_query(
    query_where_example,
    "WHERE Example: Students with Attendance Above 90"
)
```

```python
query_having_example = """
SELECT department,
       AVG(attendance_percentage) AS avg_attendance
FROM students
GROUP BY department
HAVING AVG(attendance_percentage) > 85;
"""

result_having_example = run_query(
    query_having_example,
    "HAVING Example: Departments with Average Attendance Above 85"
)
```

```python
query_department_attendance = """
SELECT department,
       AVG(attendance_percentage) AS avg_attendance
FROM students
GROUP BY department
HAVING AVG(attendance_percentage) > 85;
"""

result_department_attendance = run_query(
    query_department_attendance,
    "Query: Departments with Average Attendance Above 85%"
)
```
