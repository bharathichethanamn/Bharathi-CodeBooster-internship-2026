```python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
llm = ChatGroq(model="llama-3.3-70b-versatile")
prompt = PromptTemplate(
    input_variables=["instruction"],
    template="follow the instruction and provide the output clearly:\n\nInstruction:{instruction}\n\nAnswer"

)
parser = StrOutputParser()
chain = prompt | llm | parser
response = chain.invoke({
    "instruction": "Translate 'Good evening ' to tamil"
})
print (response)

```

```python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
llm = ChatGroq(model="llama-3.3-70b-versatile")
prompt = PromptTemplate(
    input_variables=["instruction"],
    template=(
    """
        given:day
        opposite:night
        given:large
        opposite:small
        Now, follow the instruction below:
        Instruction: {instruction}
        Answer:
     """
    )

)
parser = StrOutputParser()
chain = prompt | llm | parser
response = chain.invoke({
    "instruction": "what is the opposite of 'dark'"
})
print (response)

```
