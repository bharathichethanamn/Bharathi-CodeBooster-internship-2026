```python
from crewai import Agent, Task, Crew

def reseacrh_agent_example():
    # Define Agent
    researcher=Agent(
        role='Research Analyst',
        goal='Identify top AI trends for 2025',
        backstory='Experience technology researcher with 10+years of industry experience',
        verbose=True,
        llm='groq/llama-3.3-70b-versatile'
    )
    # Define Task
    research_task=Task(
        description='List and analyze the top 3 AI trends for 2025 with examples',
        expected_output='Comprehensive report with detailed explanation for each trend',
        agent=researcher
    )

    crew=Crew(
        agents=[researcher]
    )
```
