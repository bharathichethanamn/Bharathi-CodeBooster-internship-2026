```python
import snowflake.connector

conn = snowflake.connector.connect(
    user=SNOWFLAKE_USER,
    password=SNOWFLAKE_PASSWORD,
    account=SNOWFLAKE_ACCOUNT,
    warehouse=SNOWFLAKE_WH,
    database='INTERNSHIP_DATABASE',
    schema='ANALYTICS'
)
```

```python
def get_schema(conn):
    cursor = conn.cursor()

    cursor.execute("""
        SELECT table_name, column_name, data_type
        FROM INFORMATION_SCHEMA.COLUMNS
        ORDER BY table_name, ordinal_position
    """)

    rows = cursor.fetchall()

    schema_text = ""

    for table, column, datatype in rows:
        schema_text += f"Table: {table}, Column: {column}, Type: {datatype}\n"

    cursor.close()
    return schema_text
```

```python
def generate_sql(user_question, schema_text, client, model):

    system_prompt = f"""
    You are an expert SQL assistant.

    You are connected to a Snowflake database with the following structure:

    {schema_text}

    Rules:
    1. Generate ONLY a valid Snowflake SQL query.
    2. Return only the SQL query.
    3. No explanations.
    4. No markdown.
    5. Use only tables and columns from the schema.
    6. Use single quotes for string values.
    7. Use Snowflake SQL syntax.
    8. If the user asks for top N records, use ORDER BY and LIMIT N.
    """

    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_question}
        ],
        temperature=0.0
    )

    sql_query = response.choices[0].message.content.strip()
    return sql_query
```

```python
def execute_sql(sql_query, conn):
    try:
        cursor = conn.cursor()
        cursor.execute(sql_query)

        rows = cursor.fetchall()
        columns = [desc[0] for desc in cursor.description]

        result_df = pd.DataFrame(rows, columns=columns)

        cursor.close()

        return result_df, None

    except Exception as e:
        return None, str(e)
```

```python
def text_to_sql_agent(user_question, conn, client, model, verbose=True):

    print(f"Question: {user_question}")

    if verbose:
        print("\n[STEP 1] Reading database schema...")

    schema_text = get_schema(conn)

    if verbose:
        print("Schema loaded successfully.")

    if verbose:
        print("\n[STEP 2] Generating SQL query...")

    sql_query_text = generate_sql(
        user_question,
        schema_text,
        client,
        model
    )

    if verbose:
        print(f"Generated SQL:\n{sql_query_text}")

    if verbose:
        print("\n[STEP 3] Executing SQL query...")

    result_df, error = execute_sql(
        sql_query_text,
        conn
    )

    if error:
        print(f"SQL Execution Error: {error}")
        return None, sql_query_text

    if verbose:
        print(f"\n[STEP 4] Query returned {len(result_df)} rows")
        print("\nRESULT:")
        print(result_df.to_string(index=False))

    return result_df, sql_query_text
```
