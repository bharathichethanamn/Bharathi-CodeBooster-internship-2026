```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import chromadb
print("all package installed")
```

```python
documents=[
    "ELT is used to clean and transform data",
    "A vehicle is a mode of ttansportation",
    "Cars and trucks are popular automobiles",
    "SQL is used to query database",
    "Machine learning trains model on data",
]
query_keyword = "vehicle"
print("=" * 60)
print("KEYWORD SEARCH FOR : '{query_keyword}'")
print("=" * 60)
for i, doc in enumerate(documents):
  if query_keyword.lower() in doc.lower():
    print(f" FOUND [doc_{i}]: {doc}")
  else:
    print(f" MISSED [doc_{i}]: {doc}")
print()
print("PROBLEM: doc_2 talks about 'Cars and trucks' - which ARE vechicles")
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
print("model loaded successfully")
print("Model produces vector of size: {model.get_sentence_embedding_dimension()}:")
```

```python
sentence = "ETL is used to clean and transform data"
embedding = model.encode(sentence)
print(f"Input sentence: '{sentence}'")
print()
print(f"Embedding type:{type(embedding)}")
print(f"Embedding shape: {embedding.shape}")
print(f"Min value: {embedding.min()}")
print(f"Max value: {embedding.max()}")

```

```python
sentence = ["grow a tree",
 "grow a dog",
    "ipl over",
    "SQL is used to query database",
]
embedding = model.encode(sentence)
print(f"Number of sentence: {len(sentence)}")
print(f"shape of embedding: {embedding.shape}")
print()
print("each row is one sentence's embedding")
for i in range(len(sentence)):
    print(f"Sentence {i+1}: {sentence[i]}")
    print(f"First 5 values of embedding[{i}]: {embedding[i][:5]}")
```

```python
def cosine_smilarity(vec_a, vec_b):
   dot_product = np.dot(vec_a, vec_b)
   norm_a = np.linalg.norm(vec_a)
   norm_b = np.linalg.norm(vec_b)
   return dot_product / (norm_a * norm_b)
sim_01 = cosine_smilarity(embedding[0], embedding[1])
sim_02 = cosine_smilarity(embedding[0], embedding[2])
sim_03 = cosine_smilarity(embedding[1], embedding[2])
print(f"sentence 0: '{sentence[0]}'")
print(f"sentence 1: '{sentence[1]}'")
print(f"sentence 2: '{sentence[2]}'")
print()
print(f"Similarity (0 vs 1): {sim_01:.4f} <- Expected:HIGH (same topic)")
print(f"Similarity (0 vs 2): {sim_02:.4f} <- Expected:LOW (different topic)")
print(f"Similarity (1 vs 2): {sim_03:.4f} <- Expected:LOW (different topic)")
```
